```python

df = spark.read.option("header", "true").csv("/Workspace/Users/parthasarathy_1@msn.com/Sales.csv")
display(df)
```

```python
for col in df.columns:
    df = df.withColumnRenamed(col, col.replace(" ", "_"))
df.printSchema()

df.createOrReplaceTempView("sales_clean")

```

```python
%sql
SELECT Region, SUM(Total_Revenue) AS TotalRevenue
FROM sales_clean
GROUP BY Region
ORDER BY TotalRevenue DESC
```
